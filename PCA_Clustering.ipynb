```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
from plotnine import *

from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
from sklearn.neighbors import NearestNeighbors

from sklearn.cluster import DBSCAN


from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.cluster import AgglomerativeClustering

from yellowbrick.cluster import KElbowVisualizer

import seaborn as sns


from sklearn.metrics import silhouette_score

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def findElbowValue(model, data, metric):
    print(metric)
    visualizer = KElbowVisualizer(model, k=(2,12), metric=metric)

    visualizer.fit(data)        # Fit the data to the visualizer
    visualizer.show()        # Finalize and render the figure
    
    return visualizer.elbow_value_


def runKmeans(data):
    # optimal no of clusters
    k = findElbowValue(KMeans(), data, 'silhouette')
    
    # fitl Kmeans model with the optimal value
    km = KMeans(n_clusters=k).fit(data)
    labels = km.labels_
    score = silhouette_score(data, labels)
    
    return labels, score


def runAgglomerative(data):
    k = findElbowValue(AgglomerativeClustering(), data, 'silhouette')

    hac = AgglomerativeClustering(n_clusters=k).fit(data)
    labels = hac.labels_
    score = silhouette_score(data, labels)
#     plt.scatter(d1.x, d1.y, c=labels, cmap="rainbow")
    return labels, score


def bestGM(data):
    scores = []
    allLabels = []
    for i in range(2, 10):
        gm = GaussianMixture(n_components=i).fit(data)
        labels = gm.predict(data)
        allLabels.append(labels)

        score = silhouette_score(data, labels)
#         print(f"silhoette {i} : {score}")
        scores.append(score)
        
    k = scores.index(max(scores))+2
    
#     return k, max(scores), allLabels[k-2]
    return allLabels[k-2], max(scores)

def plotDBSCAN(data, clustering):
    
    print(data.shape)
    X= data.values
    core_samples_mask = np.zeros_like(clustering.labels_, dtype=bool)

    # set core points to True
    core_samples_mask[clustering.core_sample_indices_] = True

    labels = clustering.labels_

    plt.figure(figsize=(20, 10))
    # Black removed and is used for noise instead.
    unique_labels = set(labels)
    colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]
    for k, col in zip(unique_labels, colors):
        if k == -1:
            # Black used for noise.
            col = [0, 0, 0, 1]

        class_member_mask = labels == k

        xy = X[class_member_mask & core_samples_mask]
        plt.plot(
            xy[:, 0],
            xy[:, 1],
            "o",
            markerfacecolor=tuple(col),
            markeredgecolor="k",
            markersize=10,
        )

        xy = X[class_member_mask & ~core_samples_mask]
        plt.plot(
            xy[:, 0],
            xy[:, 1],
            "o",
            markerfacecolor=tuple(col),
            markeredgecolor="k",
            markersize=6,
        )
        
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    plt.title("Estimated number of clusters: %d" % n_clusters_)
    plt.show()
    

    
def printInfo(data, clustering):
    labels = clustering.labels_

    # Number of clusters in labels, ignoring noise if present.
    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise_ = list(labels).count(-1)
    
    print("Estimated number of clusters: %d" % n_clusters_)
    print("Estimated number of noise points: %d" % n_noise_)
    print("No of Core Points:", len(clustering.core_sample_indices_))
    print("Silhouette Coefficient: %0.3f" % silhouette_score(data, labels))
    
    
def runDBSCAN(data, eps, min_samples):
    clustering = DBSCAN(eps=eps, min_samples=min_samples).fit(data)
    printInfo(data, clustering)
    plotDBSCAN(data.iloc[:, 0:2], clustering)
```

## Handle Missing values

```python
(df.isnull().sum())
```

```python

# age, sod, hemo, rc -- mean

# bp, bgr, bu, sc, pot, pcv, wc -- median

# sg, al, su, rbc, pc  -- new category 

# pcc, ba, 'htn', 'dm', 'cad', 'appet', 'pe', 'ane' -- mode

# id drop 
```

```python
mean_col = ['age', 'sod', 'hemo', 'rc']
df[mean_col] = df[mean_col].fillna(df[mean_col].mean())

median_cols = ['bp', 'bgr', 'bu', 'sc', 'pot', 'pcv', 'wc']
df[median_cols] = df[median_cols].fillna(df[median_cols].median())

new_cols = ['sg', 'al', 'su', 'rbc', 'pc']
df[new_cols] = df[new_cols].fillna('U')

mode_cols = ['pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane' ]
df[mode_cols] = df[mode_cols].fillna(df[mode_cols].mode().iloc[0])

df = df.drop(columns=['id', 'classification'])
```

## PCA

```python
from sklearn.decomposition import PCA
 
pca = PCA(n_components = 0.95)
 
X = pca.fit_transform(df_scaled)
 
explained_variance = pca.explained_variance_ratio_
```

```python
print((explained_variance))
print(explained_variance.sum())
```

## K-Means

```python
X = pd.DataFrame(X)
labels, kmScore = runKmeans(X)

plt.figure(figsize=(20,10))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=labels, cmap="rainbow")
print("Silhouette Score: ", kmScore)
```

## Agglomerative Clustering

```python
agLabels, agScore = runAgglomerative(X)
print("Silhouette Score: ", agScore)

plt.figure(figsize=(20,10))
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=agLabels, cmap="rainbow")
```

```python
plt.figure(figsize=(20,10))
gmLabels, gmScore = bestGM(X)
print("Silhouette Score: ", gmScore)
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=gmLabels, cmap="rainbow")
```
